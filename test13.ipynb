```python
import csv
import matplotlib.pyplot as plt

f = open('seoul3.csv', 'r', encoding='UTF8')
data = csv.reader(f)
header = next(data)
jt = []
at = []
for row in data:
    if row[-1] == '' or row[-2] == '':
        continue
    if int(row[0].split('-')[0]) < 2011:
        continue
    if row[0].split('-')[1] == "01":
        row[-1] = float(row[-1])
        jt.append(row[-1])
    if row[0].split('-')[1] == "08":
        row[-1] = float(row[-1])
        at.append(row[-1])
f.close()
plt.hist(jt, bins=100, color="blue", label='Jan')
plt.hist(at, bins=100, color="red", label='Aug')
plt.legend()
plt.show
```
